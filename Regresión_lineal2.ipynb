```python
def binary(data):
  bin_reg = bin_reg = r"^[01](?:\.0)?\.?$"
  if str(data) == 'nan':
    return np.nan
  else:
    return bool(re.findall(bin_reg, str(data)))

def is_binary(df_, col):
  bd1 = bd.copy()
  percent =  bd[col].apply(binary).sum() / bd[col].count()
  if percent > 0.5:
    return True
  else:
    return False

def breakdown(bd1):
  bd1 = bd.copy()
  normal  = []
  nonormal = []
  binaries = []
  cates = []
  for col in bd.columns:
    if bd[col].dtypes.name=='int64' or bd[col].dtypes.name == 'float64':
      if is_binary(bd, col):
        binaries.append(col)
      else:
        if shapiro(bd[col]).pvalue > 0.05:
          normal.append(col)
        else:
          nonormal.append(col)
    else:
      cates.append(col)
  return normal, nonormal, binaries +  cates
```

```python
normal, nonormal, cats = breakdown(bd)
```
